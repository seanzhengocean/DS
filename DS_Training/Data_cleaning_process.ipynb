```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Data Cleaning 1st Example.csv',encoding='latin1')
df
```

```python
df.columns
```

```python
df.isnull() #df.isna
```

```python
#FIND THE COUNT OF NULL VALUES
df. isnull(). sum()
```

```python
#REPLACE NULL VALUES BY 0

df.fillna(0,inplace=True)
df
```

```python
#REPLACE NULL VALUES BY MEAN
x=df['duration'].mean()
df['duration'].fillna(x, inplace=True)
print(df)
```

```python
x=df['duration'].mean()
df['duration'].fillna(x, inplace=True)
print(df)
```

```python
for x in list(df.columns):
  try:
    df[x]=df[x].fillna(df[x].mean())
  except:
    print("couldnot fill",x)
    df[x]=df[x].fillna(0)
```

```python
df
```

```python
plt.scatter(df['num_critic_for_reviews'], df['duration'])
plt.xlabel("num_critic_for_reviews")
plt.ylabel("duration")
plt.show()
```

```python
df.to_csv('/content/Data Cleaning 1st Example.csv', mode='a', index=False, header=False)
df
```

```python
# print all even numbers for range 7
for k in range(7,8):
  allnumbers=""
  for j in range(k):
    if j%2==0:
      print(allnumbers,j)
      allnumbers=allnumbers+","+str(j)
  print("All even numbers smaller then "+str(k)+" :-",allnumbers)
```

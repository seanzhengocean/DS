```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```

```python
## Here we will check the percentage of nan values present in each feature
## 1 -step make the list of features which has missing values

features_with_na=[features for features in df_housing_price.columns if df_housing_price[features].isnull().sum()>1]
## 2- step print the feature name and the percentage of missing values
for feature in features_with_na:
    print(feature, np.round(df_housing_price[feature].isnull().mean(), 4),  ' % missing values') 
```

### Since they are many missing values, we need to find the relationship between missing values and Sales Price



Let's plot some diagram for this relationship


```python
for feature in features_with_na:
  df_housing_price_cp = df_housing_price.copy()

  # let's make a variable that indicates 1 if the observation was missing or zero otherwise
  df_housing_price_cp[feature] = np.where(df_housing_price_cp[feature].isnull(), 1, 0)

  df_housing_price_cp.groupby(feature)['SalePrice'].median().plot.bar()
  plt.title(feature)
  plt.show()
```

```python
upper_limit = df['Revenue (Millions)'].mean() + 3*df['Revenue (Millions)'].std()
lower_limit = df['Revenue (Millions)'].mean() - 3*df['Revenue (Millions)'].std()
upper_limit,lower_limit
```

```python
df1 = df.copy()
df1['Revenue (Millions)'] = np.where(df['Revenue (Millions)']>upper_limit,upper_limit,np.where(df['Revenue (Millions)']<lower_limit,lower_limit,df['Revenue (Millions)']))
```
